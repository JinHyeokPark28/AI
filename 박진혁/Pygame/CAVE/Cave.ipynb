```python
# 파이썬3 노트북 안에서 모듈 설치하는 법
# promopt -> pip install 모듈이름
# 노트북에서 pip install 모듈이름
# !pip install pygame
# 파이썬 내부의 시스템 기능을 가져옵니다.
import sys
# 난수 생성을 위한 모듈
from random import randint
import pygame
# 파이게임에 있는 우리가 사용할 기능을 임포팅
from pygame.locals import QUIT, Rect, KEYDOWN
# 파이게임이 가지고 있는 기능을 초기화합니다.
pygame.init()
# 키값에 딜레이 줘서 초기화를 시켜줍니다.
# 두개의 인자 (1 : 지연시간, 2 : 간격)
pygame.key.set_repeat(5, 5)
# 창의 이름을 바꿔줍니다.
pygame.display.set_caption('CAVE')
# 창의 크기를 설정
SURFACE = pygame.display.set_mode((800, 600))
# 프레임마다 시간 값을 가져옵니다
FPSCLOCK = pygame.time.Clock()


def main():
    # 벽이 될 변수
    walls = 80
    # 플레이어가 시작할 초기값
    ship_y = 250
    # 플레이어의 이동값
    velocity = 0
    # 점수 값
    score = 0
    # 경사 값
    slope = randint(1, 6)
    # 폰트 값
    sysfont = pygame.font.SysFont(None, 36)
    # Player_Image에 이미지를 넣습니다.
    player_image = pygame.image.load('jerry.png')
    # over_image에 이미지를 넣습니다.
    over_image = pygame.image.load('gameover.png')
    # 구멍을 가지고 있을 리스트
    holes = []
    
    # 반복문을 통해서 벽을 사각형으로 만든뒤 리스트에 집어넣습니다.
    for xpos in range(walls):
        holes.append(Rect(xpos * 10, 100, 10, 400))
    
    # 게임이 종료되었는지 확인할 변수
    game_over = False
    
    while True:
        # 스페이스바가 눌렸는지 확인하는 변수
        is_space_down = False
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                if event.key == pygame.K_SPACE:
                    is_space_down = True
        # 게임이 종료가 되지 않았다면 게임을 실행합니다.
        if not game_over:
            # 점수를 10점씩 올려준다.
            score += 10
            velocity += -3 if is_space_down else 3
            ship_y += velocity
            # 벽을 가지고 있는 리스트의 마지막 인덱스를 복사해서
            # 다른 변수에다가 넣어줍니다.
            edge = holes[-1].copy()
            # 벽을 경사만큼 움직여줍니다.
            temp = edge.move(0, slope)
            if temp.top <= 0 or temp.bottom >= 600:
                slope = randint(1, 6) * (-1 if slope > 0 else 1)
                # inflate_ip(x, y) : (x, y)의 크기 만큼 직사각형의 
                # 크기를 키우거나 줄입니다.
                edge.inflate_ip(0, -20)
            # move_ip는 직사각형을 (x, y)만큼 움직입니다.
            edge.move_ip(10, slope)
            holes.append(edge)
            del holes[0]
            holes = [x.move(-10, 0) for x in holes]
            
            if holes[0].top > ship_y or holes[0].bottom < ship_y + 80:
                game_over = True
                
            SURFACE.fill((73,111,136))
            for hole in holes:
                pygame.draw.rect(SURFACE, (0,0,0), hole)
            # 플레이어의 이미지를 배치
            
            SURFACE.blit(player_image, (0, ship_y))
            score_image = sysfont.render('Score : {}'.format(score), True, (0, 255, 0))
            SURFACE.blit(score_image, (600, 20))
            
            if game_over:
                print("게임종료")
                SURFACE.blit(over_image, (280, 180))
                
            pygame.display.update()
            FPSCLOCK.tick(15)
            

if __name__ == "__main__":
    main()
   
```
