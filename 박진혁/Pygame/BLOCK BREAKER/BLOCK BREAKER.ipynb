```python
import sys
import math
import random
import pygame
from pygame.locals import QUIT, KEYDOWN, K_LEFT, K_RIGHT, Rect

class Block:
    """블록, 공, 패들 객체"""
    def __init__(self, col, rect, speed=0):
        self.col = col
        self.rect = rect
        self.speed = speed
        # 공의 초기 위치
        self.dir = random.randint(-45, 45)
        
    def move(self):
        """공을 움직입니다."""
        self.rect.centerx += math.cos(math.radians(self.dir)) * self.speed
        self.rect.centery -= math.sin(math.radians(self.dir)) * self.speed
        
    def draw(self):
        # 속도를 기준으로 블록과 공을 구분합니다.
        if self.speed == 0:
            pygame.draw.rect(SURFACE, self.col, self.rect)
        else:
            pygame.draw.ellipse(SURFACE, self.col, self.rect)
            
def tick():
    # 각 프레임마다 프로그램이 돌 수 있도록 합니다.
    """프레임 별 처리"""
    global BLOCKS
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == KEYDOWN:
            if event.key == K_LEFT:
                PADDLE.rect.centerx -= 10
            elif event.key == K_RIGHT:
                PADDLE.rect.centerx += 10
    # 만일 볼이 y의 위치에서 1000이하에 속했있다면
    if BALL.rect.centery < 1000:
        BALL.move()
    
    # 블록과의 처리
    prevlen = len(BLOCKS)
    BLOCKS = [x for x in BLOCKS if not x.rect.colliderect(BALL.rect)]
    if len(BLOCKS) != prevlen:
        BALL.dir *= -1
        
    # 패들과의 처리 
    if PADDLE.rect.colliderect(BALL.rect):
        # 이 게임이 다 완성된 후 숫자 바꿔보기
        BALL.dir = 90 + (PADDLE.rect.centerx - BALL.rect.centerx) / PADDLE.rect.width * 80
    
    # 벽과의 충돌
    # BALL.dir의 값을 바꿔보기
    if BALL.rect.centerx < 0 or BALL.rect.centerx > 600:
        BALL.dir = 180 - BALL.dir
    if BALL.rect.centery < 0:
        BALL.dir = -BALL.dir
        BALL.speed = 15
        
pygame.init()
pygame.key.set_repeat(5, 5)
SURFACE = pygame.display.set_mode((600, 800))
FPSCLOCK = pygame.time.Clock()
BLOCKS = []
PADDLE = Block((242, 242, 0), Rect(300, 700, 100, 30))
BALL = Block((242, 100, 0), Rect(300,400, 20, 20), 10)

def main():
    myfont = pygame.font.SysFont(None, 92)
    mess_clear = myfont.render("CLEARED!!", True, (180, 243, 96))
    mess_over = myfont.render("GAME OVER!!", True, (180, 243, 96))
    fps = 30
    colors = [(0,255,0), (164, 255, 0), (155,155,155), (0,128,0), (242, 242, 242), (0, 0, 130)]
    
    for ypos, color in enumerate(colors, start=0):
        for xpos in range(0, 5):
            BLOCKS.append(Block(color, Rect(xpos* 100 + 60, ypos * 50 + 40, 80, 30)))
    
    while True:
        tick()
        
        SURFACE.fill((0,0,0))
        BALL.draw()
        PADDLE.draw()
        for block in BLOCKS:
            block.draw()
            
        if len(BLOCKS) == 0:
            SURFACE.blit(mess_clear, (120, 300))
        if BALL.rect.centery > 800 and len(BLOCKS) > 0:
            SURFACE.blit(mess_over, (120, 300))
        pygame.display.update()
        FPSCLOCK.tick(fps)
        
if __name__ == "__main__":
    main()
```
