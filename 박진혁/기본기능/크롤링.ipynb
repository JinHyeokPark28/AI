```python
# 모듈 : 크롤링에 필요한 모듈
# requests
# bs4.BeautifulSoup
import requests
from bs4 import BeautifulSoup
res = requests.get('https://post.naver.com/viewer/postView.nhn?volumeNo=30601579&memberNo=41596968')
soup = BeautifulSoup(res.content, 'html.parser')
# 태그값 : html 태그를 통해서 데이터를 나눠주기 때문에
# 태그를 통해서 원하는 값을 찾아 줄수 있습니다.
data = soup.find('span')
print(data.get_text())
```

```python
html = "<html>  \
            <body>\
                <h1 id='title'>[1]크롤링은 쉬울까요?</h1> \
                <p class='cssstyle'> 웹페이지를 훔쳐봅시다!</p> \
            </body> \
        </html>"
# 정보를 가져와보세요
soup = BeautifulSoup(html, 'html.parser')
# h1 텍스만 가져와보세요.
data = soup.find('h1')
print(data.get_text())
# 간혹 get_text() 실행시켰는데 None이 나오는 경우
# 혹은 아무것도 나오지 않은 경우에는
# print(data.string)
print(data.string)
```

```python
html = "<html>  \
            <body>\
                <h1 id='title'>[1]크롤링은 쉬울까요?</h1> \
                <p class='cssstyle'> 웹페이지를 훔쳐봅시다!</p> \
                <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술이 발달되어 있습니다.</p> \
            </body> \
        </html>"
soup = BeautifulSoup(html, 'html.parser')
data=soup.find('p', class_='cssstyle')
print(data.string)
data = soup.find('p', attrs={'align':'center'})
print(data.string)
```

```python
# for문을 통해서 여러번 나눠서 실행시켜줘야한다.
data = soup.find_all('p')
for item in data:
    print(item.get_text())
```

```python
# Selector 형식의 속성
# css selector의 기능으로써
# html을 선택해서 tag, tag+class, 여러개의 클래스를 보다 쉽게
# 정보를 가져오기 위해서 사용됩니다.
# soup.find() => soup.select()
res = requests.get("https://comic.naver.com/index.nhn")
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('div.genreRecomInfo2 > h6 > a > span')
for item in data:
    print(item.get_text())
```

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://tv.naver.com/r')
raw = res.text
soup = BeautifulSoup(raw, 'html.parser')
data = soup.select('div.cds')
text = open('TOP100.txt', 'w', encoding='utf-8')
for item in data:
    title = item.select_one('tooltip').text
    hit = item.select_one('span.hit').text
    text.write(title)
    text.write('  /  ')
    text.write(hit)
    text.write('\n')
text.close()
```
